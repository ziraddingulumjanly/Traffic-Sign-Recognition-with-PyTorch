## Mathematical view

The classifier learns

$$
f_\theta(x) = \hat{y}, \qquad \hat{y} \in \{0, 1, \dots, 42\}
$$

where $x$ is a traffic-sign image and $\hat{y}$ is the predicted sign class.


```python
from torchvision.datasets import GTSRB

from gtsrb_classifier.config import CLASS_NAMES
from gtsrb_classifier.data import build_transforms
from gtsrb_classifier.plots import plot_class_distribution

train_dataset = GTSRB(
    root="../data",
    split="train",
    download=True,
    transform=build_transforms(image_size=64, split="eval"),
)

len(train_dataset), len(CLASS_NAMES)
```

```python
labels = [label for _, label in train_dataset._samples]
plot_class_distribution(
    labels=labels,
    class_names=CLASS_NAMES,
    save_path="../reports/figures/class_distribution.png",
)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 5, figsize=(10, 6))
for ax, index in zip(axes.ravel(), range(15), strict=False):
    image, label = train_dataset[index]
    ax.imshow(image.permute(1, 2, 0))
    ax.set_title(f"{label}: {CLASS_NAMES[label][:18]}", fontsize=7)
    ax.axis("off")
fig.tight_layout()
```
